```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- Configurações Iniciais (Garanta que estas bibliotecas foram importadas no início do seu notebook) ---
plt.style.use('ggplot')
sns.set_theme(style="whitegrid")

# --- Defina o caminho da sua pasta de imagens ---
caminho_imagens = r'D:\03 - Portfolios\portfolio_diego\analise_reinfaction_covid\imagens'
os.makedirs(caminho_imagens, exist_ok=True) # Cria a pasta se ela não existir

# --- Carregue seu DataFrame (se não estiver já carregado na mesma sessão) ---
# Se o seu DataFrame 'df' já estiver carregado e pronto para uso das etapas anteriores,
# você pode pular esta parte. Caso contrário, execute-a:
caminho_csv = r'D:\03 - Portfolios\portfolio_diego\analise_reinfaction_covid\dados\covid_related_disease_data.csv'
try:
    df = pd.read_csv(caminho_csv)
    print("DataFrame carregado para geração de gráficos.")
except Exception as e:
    print(f"Erro ao carregar o DataFrame: {e}. Certifique-se que 'df' está disponível.")


# --- Geração dos Gráficos com os NOMES CORRETOS das Colunas ---

print("\n--- Gerando gráficos... ---")

# Gráfico 1: Histograma da Idade
if 'Age' in df.columns and pd.api.types.is_numeric_dtype(df['Age']):
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Age'].dropna(), bins=30, kde=True)
    plt.title('Distribuição da Idade dos Pacientes')
    plt.xlabel('Idade (Age)')
    plt.ylabel('Frequência')
    plt.grid(True)
    plt.savefig(os.path.join(caminho_imagens, 'distribuicao_idade.png'))
    plt.show()
    print("Gráfico 'distribuicao_idade.png' gerado.")
else:
    print("Coluna 'Age' não encontrada ou não é numérica para Histograma da Idade.")


# Gráfico 2: Contagem por Gênero
if 'Gender' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Gender', data=df.dropna(subset=['Gender']), palette='viridis')
    plt.title('Contagem de Pacientes por Gênero')
    plt.xlabel('Gênero (Gender)')
    plt.ylabel('Número de Pacientes')
    plt.savefig(os.path.join(caminho_imagens, 'contagem_genero.png'))
    plt.show()
    print("Gráfico 'contagem_genero.png' gerado.")
else:
    print("Coluna 'Gender' não encontrada para Contagem por Gênero.")


# Gráfico 3: Relação entre Idade e Severidade dos Sintomas (Scatter Plot)
if 'Age' in df.columns and 'Severity' in df.columns and \
   pd.api.types.is_numeric_dtype(df['Age']) and pd.api.types.is_numeric_dtype(df['Severity']):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Age', y='Severity', data=df.dropna(subset=['Age', 'Severity']))
    plt.title('Idade vs. Severidade dos Sintomas')
    plt.xlabel('Idade (Age)')
    plt.ylabel('Severidade dos Sintomas (Severity)')
    plt.savefig(os.path.join(caminho_imagens, 'idade_vs_severidade.png'))
    plt.show()
    print("Gráfico 'idade_vs_severidade.png' gerado.")
else:
    print("Colunas 'Age' ou 'Severity' não encontradas/numéricas para Scatter Plot.")


# Gráfico 4: Casos de Reinfecção por Condição Preexistente (Count Plot com Hue)
# Assumindo 'Reinfection' é booleano/binário (ex: True/False, 0/1) ou categórico 'Yes'/'No'
if 'Reinfection' in df.columns and 'Preexisting_Condition' in df.columns:
    plt.figure(figsize=(12, 7))
    sns.countplot(x='Preexisting_Condition', hue='Reinfection', data=df.dropna(subset=['Reinfection', 'Preexisting_Condition']), palette='coolwarm')
    plt.title('Casos de Reinfecção por Condição Preexistente')
    plt.xlabel('Condição Preexistente (Preexisting_Condition)')
    plt.ylabel('Número de Casos')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Reinfecção')
    plt.tight_layout()
    plt.savefig(os.path.join(caminho_imagens, 'reinf_por_condicao_preexistente.png'))
    plt.show()
    print("Gráfico 'reinf_por_condicao_preexistente.png' gerado.")
else:
    print("Colunas 'Reinfection' ou 'Preexisting_Condition' não encontradas para Reinfecção por Condição Preexistente.")


# Gráfico 5: Mapa de Calor da Matriz de Correlação
# Este gráfico pegará todas as colunas numéricas automaticamente.
numeric_cols = df.select_dtypes(include=['number']).columns
if len(numeric_cols) > 1:
    plt.figure(figsize=(12, 10))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('Mapa de Calor da Matriz de Correlação')
    plt.savefig(os.path.join(caminho_imagens, 'heatmap_correlacao.png'))
    plt.show()
    print("Gráfico 'heatmap_correlacao.png' gerado.")
else:
    print("Não há colunas numéricas suficientes para gerar o Mapa de Calor da Correlação.")


# Gráfico 6: Distribuição do Intervalo entre Infecções (Histograma)
# Primeiro, certifique-se de converter as colunas de data para o tipo datetime.
if 'Date_of_Infection' in df.columns and 'Date_of_Reinfection' in df.columns:
    # Conversão para datetime com 'errors=coerce' para lidar com possíveis formatos inválidos
    df['Date_of_Infection'] = pd.to_datetime(df['Date_of_Infection'], errors='coerce')
    df['Date_of_Reinfection'] = pd.to_datetime(df['Date_of_Reinfection'], errors='coerce')

    # Filtra apenas casos onde houve reinfecção (Reinfection == True ou 'Yes')
    # Ajuste 'True' ou 'Yes' conforme o valor real na sua coluna 'Reinfection'
    df_reinfected = df[df['Reinfection'] == True].copy() # Ou df[df['Reinfection'] == 'Yes'].copy()

    # Calcula o intervalo apenas para casos de reinfecção e onde ambas as datas são válidas
    if not df_reinfected.empty:
        df_reinfected['intervalo_dias'] = (df_reinfected['Date_of_Reinfection'] - df_reinfected['Date_of_Infection']).dt.days

        plt.figure(figsize=(10, 6))
        sns.histplot(df_reinfected['intervalo_dias'].dropna(), bins=30, kde=True)
        plt.title('Distribuição do Intervalo (dias) entre Infecções (Casos de Reinfecção)')
        plt.xlabel('Dias entre Infecções')
        plt.ylabel('Frequência')
        plt.savefig(os.path.join(caminho_imagens, 'intervalo_reinfecao.png'))
        plt.show()
        print("Gráfico 'intervalo_reinfecao.png' gerado.")
    else:
        print("Nenhum caso de reinfecção encontrado ou datas inválidas para o gráfico de intervalo.")
else:
    print("Colunas de datas de infecção ('Date_of_Infection' ou 'Date_of_Reinfection') não encontradas para o gráfico de intervalo.")

print("\n--- Geração de gráficos concluída! Verifique sua pasta de imagens. ---")
```
